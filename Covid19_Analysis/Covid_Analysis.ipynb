```python
# Removing unwanted columns

covid_df.drop(['Sno', 'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Time'], inplace=True, axis=1)
```

```python
covid_df.head(10)
```

```python
# Changing Date format using pandas inbuilt function

covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
covid_df.head(10)
```

```python
# Active cases

covid_df['Active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values=['Confirmed', 'Deaths', 'Cured'], index='State/UnionTerritory', aggfunc=max)
```

```python
statewise['Recovery Rate'] = statewise['Cured'] * 100/statewise['Confirmed']
statewise['Mortality Rate'] = statewise['Deaths'] * 100/statewise['Confirmed']
statewise = statewise.sort_values(by='Confirmed', ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
# Top 10 active cases states

top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by=['Active_cases'], ascending=False).reset_index()
```

```python
top_10_active_cases = covid_df.groupby(by='State/UnionTerritory').max()[['Active_cases', 'Date']].sort_values(by=['Active_cases'], ascending=False).reset_index()
fig = plt.figure(figsize=(16, 9))
plt.title("Top 10 states with most active cases in India", size = 25)
ax = sns.barplot(data= top_10_active_cases.iloc[:10], y="Active_cases", x='State/UnionTerritory', linewidth = 1, edgecolor='black')
plt.xlabel('States')
plt.ylabel('Total Active Cases')
plt.show()
```

```python
# Top states with highest deaths

top_10_deaths = covid_df.groupby(by='State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by=['Deaths'], ascending=False).reset_index()
```

```python
# Growth trend

fig = plt.figure(figsize=(14, 8))
ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerla', 'Tamil Nadu', 'Uttar Pradesh'])], x = 'Date', y = 'Active_cases', hue = 'State/UnionTerritory')
ax.set_title("Top 5 Affected states in India", size = 16)
```

```python
# Remove rows where state = India

vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated' : 'Total'}, inplace=True)
vaccine
```

```python
# States with most number of vaccinated Individuals 

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending=False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(15, 14))
plt.title("Top 5 Vaccinated States", size = 20)
x = sns.barplot(data= max_vac.iloc[:10], y=max_vac.Total, x=max_vac.index, linewidth = 1, edgecolor='black')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

```python
# States with least number of vaccinations

least_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
least_vac = least_vac.sort_values('Total', ascending=True)[:5]
least_vac
```

```python
fig = plt.figure(figsize=(15, 14))
plt.title("Top 5 Least Vaccinated States", size = 20)
x = sns.barplot(data= least_vac.iloc[:10], y=least_vac.Total, x=least_vac.index, linewidth = 1, edgecolor='black')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
